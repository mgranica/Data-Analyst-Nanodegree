```python
# import and load data
import pandas as pd
%matplotlib inline

df = pd.read_csv('powerplant_data_edited.csv')
df.head()
```

```python
# plot relationship between temperature and electrical output
df.plot(x='Temperature', y=' Energy Output', kind='scatter', figsize=(16,9) );
```

```python
# plot distribution of humidity
df['Relative Humidity'].hist(figsize=(16,9));
```
