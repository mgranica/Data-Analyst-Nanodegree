```python
# imports and load data
import pandas as pd
% matplotlib inline
```

```python
df = pd.read_csv('store_data.csv')
df.head()
```

```python
# explore data
df.shape
```

```python
df.info()
```

```python
df.describe().T
```

```python
df['week'] = pd.to_datetime(df['week'])
df.tail()
```

```python
# sales for the last month
mask = df.week > '2018-02-01'
ax = df.loc[mask].sum().plot(figsize=(16,9), kind='bar');
ax.set_title('sales for the last month', fontsize=20)
ax.set_xlabel('stores', fontsize=18)
ax.set_ylabel('sales',fontsize=18)
ax.tick_params(labelsize=15)
```

```python
# average sales

ax= df.mean().plot(figsize=(16,9), kind='bar');
ax.set_title('average sales', fontsize=20)
ax.set_xlabel('stores', fontsize=18)
ax.set_ylabel('sales', fontsize=18)
ax.tick_params(labelsize=15)
```

```python
# sales for the week of March 13th, 2016
df.loc[df.week == '2016-03-13'].plot(figsize=(16,9), kind='bar');
```

```python
# sales for the latest 3-month periods
ax=df.loc[df.week>'2017-12-01'].sum().plot(figsize=(16,9), kind='bar');
ax.set_title('sales for the latest 3-month periods', fontsize=20)
ax.set_xlabel('stores', fontsize=18)
ax.set_ylabel('sales', fontsize=18)
ax.tick_params(labelsize=15)
```
